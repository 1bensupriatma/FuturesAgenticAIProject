# Agentic AI for Two-Bar Fibonacci Retrace Strategy (Futures)


## 1. Objective & Design Brief


Build an educational analysis agent for a Two-Bar Fibonacci Retrace futures strategy. The project detects deterministic trade setups, prepares structured data, and uses an LLM only to explain results.

## 2. Strategy Explanation

- Use the last two completed bars as the impulse.
- Require each bar body to be at least 50% of total range.
- Require Bar 2 volume to be greater than Bar 1 volume.
- Classify matched bullish, matched bearish, mixed bullish, or mixed bearish impulses.
- Compute the 50% to 61.8% retracement zone from Bar 2 only.
- Use the current bar as the entry trigger bar.
- Apply optional VWAP alignment before producing a trade plan.

## 3. Agentic Workflow Explanation

1. Load CSV data.
2. Validate OHLCV columns.
3. Compute VWAP.
4. Detect the two-bar impulse.
5. Compute the retracement zone.
6. Validate the entry trigger.
7. Apply the VWAP filter.
8. Generate a structured trade plan.
9. Optionally run a starter backtest.
10. Send structured output to an LLM for explanation only.

## 4. Prompt Iteration Log

- Initial weak prompt: asks the LLM to explain the setup with little control.
- Improved prompt: tells the LLM not to invent signals or recalculate values.
- Final structured prompt: requires fixed output fields and keeps logic in Python.
